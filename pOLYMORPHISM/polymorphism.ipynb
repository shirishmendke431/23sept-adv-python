### create class Reactangle ,circle --> perimeter , area

```python
from dataclasses import dataclass
```

```python
@dataclass
class Reactangle:
    s1 : int|float
    s2 : int|float

# perimeter
    def Perimeter(self):
        return 2*(self.s1+self.s2)
    
# area
    def Area(self):
        return self.s1*self.s2
    
# create class circle
@dataclass
class Circle:
    r : int|float
#perimeter
    def Perimeter(self):
        return 2*3.14*self.r
#area 
    def Area(self):
        return 3.14*self.r**2
```

```python
R1 = Reactangle(12,32)
```

```python
R1.Area()
```

```python
R1.Perimeter()
```

```python
c1 = Circle(2)
```

```python
c1.Area()
```

```python
c1.Perimeter()
```

```python
c2 = Circle(3)
```

```python
c2.Area()
```

```python
c2.Perimeter()
```

```python
@dataclass
class  Triangle:
    s1 :  int|float
    s2  :  int|float
    s3 :  int|float

#  perimeter
    def  Perimeter(self):
        if  (self.s1 + self.s2 >  self.s3) and \
        (self.s2 + self.s3> self.s1)  and  \
        (self.s1 + self.s3  >  self.s2) : 
            P  =  self.s1+self.s2+self.s3
            return P
        

 #  find area  of  triangle
    def  Area(self):
        if self.Perimeter() is not None:
            s = self.Perimeter()/2
            return  (s*((s-self.s1)*(s-self.s2)*(s-self.s3)))**(1/2)
        else :
            print('sides do not formed triangle')
   
```

```python
t1 = Triangle(s1 = 12,s2=13,s3=23)
```

```python
t1.Perimeter()
```

```python
t1.Area()
```

```python
for i in [t1,R1,c1]:
    print(i)
    print(f'Perimeter --> {i.Perimeter()}')
    print(f'Area --> {i.Area()}')
    print()
    print('---------------------')
```

# Encapsulation
- Public variable - access publically
- protected variable - _
- private variables - __

```python
# Public variable
class Account:
    def __init__(self,bal):
        self.bal= bal

```

```python
a = Account(20000)
```

```python
a.bal
```

```python
# Protected variable
class Account1:
    def __init__(self,bal):
        self._bal= bal
# check balance
    def check_balance(self):
        print(self._bal)
```

```python
a1 = Account1(200)
```

```python
a1._bal
```

```python
a1.check_balance()
```

```python
# private variables
class Account2:
    def __init__(self,bal):
        self.__bal = bal
```

```python
a2 = Account2(300)
```

```python
a2
```

```python
a2.__bal()
```

```python
a2._Account2__bal
```

### classmethod

```python
class Student:
    school_name  =  'abc'
    def  __init__(self  , name  , age):
        self.name  =  name
        self.age  = age 

    @classmethod
    def  change_school(cls ,  new_name) :
        cls.school_name  =  new_name 

    # 
    @classmethod
    def school(cls):
        print(f'name  of school is  {cls.school_name}')
```
