# decorators

```python
#  create decorator  to welcome  the  user
def  Welcome(func):
  def wrapper(*args  , **kwargs):
    print('Welcome user')
    result  =  func(*args  , **kwargs)
    print(f'Result  of  function is {result}')
    print('Thank  you')

    return result
  return  wrapper

```

```python
#  create  function  for  factorial  of  number & apply decorator
@Welcome
def  factorial(num):
  fact  =  1
  for i  in range(1 , num+1):
    fact=  fact*i
  return  fact
```

```python
factorial(5)
```

```python
@Welcome
def square(num):
    return num**2
```

```python
square(44)
```

```python
@Welcome
def  hotel_bill(**kwargs):
  menu  =  {'Pizza' :  299 ,  "Burgur" :  100  , "Sandwich" :  60}
  total  = 0
  for  food  ,  quantity  in  kwargs.items():
    if  menu.get(food.title()) is  not  None:
      price  =  menu[food.title()]*quantity
      print(f' total  of {food} is {price} ')
      total  = total+price
  return f"Total  price is {total}"
```

```python
hotel_bill(pizza = 2 , samosa = 3)
```

```python
square(num= 3)
```

## create decorator  for authentication

```python
user  =  {"sarthak1" :  'sarthak@123' ,
          'sunita_a' :  'sunita@90'}
```

```python
def Authentication(func):
  def wrapper(username  ,  password  , *args  , **kwargs):
    if username in user.keys() and  password == user[username]:
      print('Login Successful')
      result  =  func(username  , password  , *args  , **kwargs)
      print(f'Output of  function is  {result}')
      print(f'Thank you {username}')

    else:
      raise ValueError('Please  check  login credentials')
  return  wrapper
```

```python
## create  function to  find  simple  interest
@Authentication
def  simple_interest(username  , password  ,P: int |float  ,  N : int , R : int|float):
  """
  p:Principal amount
  N : number of  years
  R : rate  of interest
  """
  I  =  P*R*N/100
  A = P+I
  return  f"Interest --> {I} ,  Amount  to  be  return  {A}"

```

```python
## pass data P =  100000  , N  = 2 , R  = 6.5
simple_interest('sarthak1' , 'sarthak@123' ,P =  100000 , N =  2 , R  =  6.5)
```

```python
# create decorator  to  find elapsed time  to executes  output  of function
```

```python
import time
```

```python
def elapsed_time(func):
  def wrapper(*args  , **kwargs):
    import time
    start  =  time.time()
    result  =  func(*args  , **kwargs)
    print(f'output of function  is {result} ')
    end  =  time.time()
    print('elapsed time  is ' , round((end  - start),  6))
  return  wrapper
```

```python
# create function  to find AREA  OF  CIRCLE
@elapsed_time
def  circle_area(r):
  return 3.14*r**2
```

```python
circle_area(2)
```

```python
#  create  function  to find number  is prime  number or not
@elapsed_time
def prime(num):
  for  i  in range(2 , num):
    if num%i == 0:
      print(f'{num} is  not prime')
      break
  else:
    print(f'{num} is  prime')
```

```python
prime(22)
```
