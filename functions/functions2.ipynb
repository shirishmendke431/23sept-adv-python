```python
def info(**kwargs):
    user = { "name": None,
            "age": None,
            "email": None,
            "gender": None  }
    user.update(kwargs)
    return user
```

```python
def info(**kwargs):
    user = { "name": None,
            "age": None,
            "email": None,
            "gender": None  }
    for key in kwargs.keys():
        if key in user.keys():
            user[key]= kwargs[key]
    return user
```

```python
def info(**kwargs):
    user = { "Name": None,
            "Age": None,
            "Email": None,
            "Gender": None  }
    for key in kwargs.keys():
        if key.title() in user.keys():
            user[key.title()]= kwargs[key]
    return user
```
