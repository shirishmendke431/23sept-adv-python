### create function to add only numbers

```python
def add1(x: int|float , y: int|float):
    """
    add two numbers

    """
    if isinstance(x , (int|float)) and  \
       isinstance( y,(int|float)):
       z = x + y
       return z
    else :
        print("data is not numeric")
```

```python
a = add1(x = 'python' ,  y = '34')
a
```

```python
b =add1 (x=23,y=23)
b
```

```python
c = add1([12,34],[56,8,89])
c
```

### use assert to raise error

```python
def add2(x: int|float , y: int|float):
    """
    add two numbers

    """
    assert isinstance(x,(int,float)) , "x should be numeric"
    assert isinstance(y,(int,float)) , "y should be numeric"
    z = x+y
    return z
```

```python
add2(2,3)
```

```python
add2('12',21)
```

```python
add2(23,'34')
```

```python
add2(12,34,435)
```

### variadic function
-- varing arguments function

- variable length arguments - * 
- key word variable name arguments - **

```python
def args_len(*args):
    print(args)
    print(type(args))
```

```python
args_len(12,34,56,56)
```

```python
def add4(*args):
    add = 0
    for i in args:
        if isinstance (i ,(int,float)):
            add =  add+i
    return add
```

```python
add4(12,'34','45',456)
```

```python
add4(12,3,4,6)
```

### list unpacking

```python
l = [ 12,34,4,7]
```

```python
def add4(*args):
    add = 0
    for i in args:
        add =  add+i
    return add
```

```python
add4(l)
```

```python
print(l)
add4(*l)
```

```python
def total_bill(*args):
    total = 0 
    bill_before_tax = 0
    for product , quantity , price in args:
        subtotal = price*quantity
        print(f'sub total of {product}--> {subtotal}')

        bill_before_tax =  subtotal + bill_before_tax
        tax = 0.18
        taxable_amount = bill_before_tax*0.18

        total = taxable_amount + bill_before_tax
    print(f'bill_before_tax --> {bill_before_tax}')
    print(f'taxable_amount -->{taxable_amount}')
    print(f'total_amount -->{total}')
```

```python
total_bill(('mobile',2 , 30000))
```

```python
total_bill(('mobile',3,20000),('fan',2,2000))
```

```python
def total_bill_with_discount(*args):
    total = 0 
    bill_before_tax = 0
    for product , quantity , price, discount in args:
        subtotal = price*quantity - price*quantity*discount/100
        print(f'sub total of {product}--> {subtotal}')

        bill_before_tax =  subtotal + bill_before_tax
        tax = 0.18
        taxable_amount = bill_before_tax*0.18

        total = taxable_amount + bill_before_tax
    print(f'bill_before_tax --> {bill_before_tax}')
    print(f'taxable_amount -->{taxable_amount}')
    print(f'total_amount -->{total}')
```

```python
total_bill_with_discount(('mobile',1,30000,10))
```

```python
total_bill_with_discount(('mobile',1,30000,10),('earphone',2,1500,20))
```

```python
total_bill_with_discount(('mobile',1,30000,10),('earphone',2,1500,20),('laptop',2,25000,10))
```
