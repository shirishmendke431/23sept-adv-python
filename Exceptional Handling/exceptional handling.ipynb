### use try and except block to handle division
#### create Function

```python
def  div1():
    try:
        num =  float(input('Enter number here'))
        den =  float(input('enter number here'))
        div  =  num/den
        print (f'division of {num} and {den} --> {num}/{den}')
    
        return div

    except Exception as e:
        print(f'Exception occured at {e}')
```

```python
# take num = 200 , den = 20
a =div1()
a
```

```python
# take num = 20 , den = 0
div1()
print('hello')
```

```python
#take num = 230 , den = '12'\
div1()
for i in range(1,11):
    print(i)
```

### use customized message for division 

```python
def  div2():
    try:
        num =  float(input('Enter number here'))
        den =  float(input('enter number here'))
        div  =  num/den
        print (f'division of {num} and {den} --> {num}/{den}')
    
        return div

    except ZeroDivisionError:
        print('Denominator should not be zero')
    except ValueError:
        print('Input should not be string or blank')
    except Exception as e:
        print(f'Exception occured at {e}')
```

```python
# take num =12 ,den =6
div2()
```

```python
# take den = 0
div2()
```

```python
# take num = '12' , den = 6
div2()
```

### handle file reading exceptional cases

```python
def file_read(path):
    import pandas as pd
    try:
        df = pd.read_excel(path)
        return df 
    except FileNotFoundError:
        print('please check file or path')
```

## Four blocks
- try : checking condition
- except : when try is false
- else : it will executes when try is true
- finally : it will always executes

```python
# create function to find square of number
def square():
    num = float(input())
    return num*num
    
```

```python
square()
```

```python
def square():
    try:
        num = float(input())
        s = num*num
    except Exception as e:
        print(e)
    else:
        print(f'square of {num} is {num}*{num}')
        return s
    finally:
        print('i will always executes')
```

```python
# num = 3
square()
```

```python
# take num = '1'
square()
```
