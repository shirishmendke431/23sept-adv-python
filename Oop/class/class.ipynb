### create class account with 
check balance , deposite , withdraw , transfer methods number , name , balance --attr

```python
@dataclass
class Acount:
    acc_no  :  int 
    acc_name  : str
    balance  :  float
## create  method check balance

    def check_balance(self):
        return f'Current  balance  of  acount {self.acc_no} is {self.balance}'
    
 ## add  money to acount  deposite  `
    def Deposite(self  , amt : float):
        self.balance  +=amt
        print(f'Amount  of {amt} is deposited in  {self.acc_no}')
        return self.check_balance()
## withdraw  amount

    def Withdraw(self  , amt):
        if self.balance>=amt:
            self.balance -= amt
            print(f'Amount of  {amt} is  withdrawn from {self.acc_no}')
            return  self.check_balance()
        else:
            raise ValueError('Insufficient funds ')

   #3 transfer money 
    def  Transfer(self ,    amt  : float , acc =  'Acount' ):
        self.Withdraw(amt)
        acc.Deposite(amt)
        return  self.check_balance() , acc.check_balance()
```

create class to find area and perimeter of triangle


![download (1).png](<attachment:download (1).png>)

- s1 , s2 , s3
- s1+s2>s3 and s1+s3>s2 and s2+s3 > s1
- p = perimeter
- s = p/2

```python
s1 = 213
s2 = 341
s3 = 45
if s1+s2>s3 and s1+s3>s2 and s2+s3>s3:
    print('triangle is formed')
else:
    print('triangle is not formed')
```

```python
# create class triangle
from dataclasses import dataclass 
```

```python
@dataclass
class  Triangle:
    s1 :  int|float
    s2  :  int|float
    s3 :  int|float

#  perimeter
    def  Perimeter(self):
        if  (self.s1 + self.s2 >  self.s3) and \
        (self.s2 + self.s3> self.s1)  and  \
        (self.s1 + self.s3  >  self.s2) : 
            P  =  self.s1+self.s2+self.s3
            return P
        else :
            print('Triangle  is  not formed')
    

    #  find area  of  triangle
    def  Area(self):
        s  =  self.Perimeter()
        return  (s*((s-self.s1)*(s-self.s2)*(s-self.s3)))**(1/2)
```

```python
@dataclass
class  Triangle:
    s1 :  int|float
    s2  :  int|float
    s3 :  int|float

#  perimeter
    def  Perimeter(self):
        if  (self.s1 + self.s2 >  self.s3) and \
        (self.s2 + self.s3> self.s1)  and  \
        (self.s1 + self.s3  >  self.s2) : 
            P  =  self.s1+self.s2+self.s3
            return P
        

 #  find area  of  triangle
    def  Area(self):
        if self.Perimeter() is not None:
            s = self.Perimeter()/2
            return  (s*((s-self.s1)*(s-self.s2)*(s-self.s3)))**(1/2)
        else :
            print('sides do not formed triangle')
   
```

#### create class for simple calcultor
two attributes
add, substraction , producr ,division 

```python
@dataclass
class  Calculator:
    x :  int|float
    y :  int|float

    # addition  method
    def add(self):
        if  isinstance(self.x , (int , float)) and \
        isinstance(self.y , (int , float)):
            return  self.x+self.y
    
    # subtraction
    def subtraction(self):
        if self.add() is  not None:
            return  self.x-self.y
    
    #  product
    def  product(self):
        if self.add()is not None:
            return self.x*self.y
        
    #  def division
    def div(self):
        if  self.add() is  not None and  self.y !=0:
            return self.x/self.y
        else:
            raise TypeError('denominator should  not be 0 '
            'or input should  be numeric')
```

```python
c1 = Calculator(x = 12, y = 23)
```

```python
print(c1.add() ,  c1.div() , c1.subtraction() , c1.product())
```
