```python

## create funct5ion for  hypotenus  of  triangle
def  Hypotenus(s1  ,  s2):
  print('hypotenus  function is  started')
  time.sleep(2)
  h  =  (s1*s1+s2*s2)**(1/2)

  print(f'Hypotenus -- >  {h}')
  print('Hypotenus function  is completed')
```

```python
# create  function  to  find simple interest
def  simple_interest(P :  int|float  ,  N : int|float  ,  R : int|float):
  print('Simple interest  function is  started')
  '''
  P : principal  amount
  R : Rate  of interest
  N : no  of  years
  '''
  time.sleep(3)
  I  =  P*N*R/100
  A  =  P + I
  print(f'interest  :  {I}  , Amount  to be  return  :  {A}' )
  print('Simple ineterst  function id completed')
```

```python
%%time
simple_interest(100000  ,  2  ,  6.5)
```

```python
%%time
Hypotenus(s1  =  45  , s2  =  56)
simple_interest(P  = 200000 , R  =  7  ,  N =  3)
```

```python
# create multiple threads to  start all functions  at the  same  time
```

```python
from threading  import Thread
```

```python

%%time
th1  =  Thread(target  =  Hypotenus ,  args = (   22 , 34) )
th2  =  Thread(target  =  simple_interest , args =  (  200000 , 4  , 3 ))

#  start  threads
th1.start()
th2.start()

#  use  join
th1.join()
th2.join()
```

```python
#  create function to  find  factorial  of  number
def factorial(n):
  print('factorial  function is started')
  time.sleep(1)
  fact  =  1
  for  i  in range(1 ,n+1 ):
    fact =  fact*i
  print(f'factorial of  {n} -->  {fact}')
```

```python
#  print numbers  from 1  to 5
for  i  in range(1 , 6):
  print(i)
```

```python
#  print nubers  from  10 to 1
for  i  in range(10 , 0 , -1):
  print(i , end  =  '  ')
```

```python
%%time
factorial(8)
```

```python
%%time
l  =  [5,7,12 ,18,4]
for  i in l:
  factorial(i)
```

```python
def  multithread(l):
  thread  =  []
  for i  in l:
    th = Thread(target  =  factorial ,  args  =  (i , ))
    th.start()

    thread.append(th)
  #  join all thraeds  which are in thread
  for  thr in thread:
    thr.join()
```

```python
%%time
multithread( l  = [9 ,4,6,12,7])
```

```python
from urllib.request import urlretrieve
```

```python

def  file_download(url):

  file_name  = url.split('/')[-1]
  print(f'{file_name} is  started  to download')
  urlretrieve(url ,  file_name)
  print(f'{file_name} is  downloaded')
```

```python
#  create  file_download_multithread function to start downloading at the same  time
def  multi_file_download(l):
   thread  =  []
   for i  in l:
    th = Thread(target  =  file_download ,  args  =  (i , ))
    th.start()

    thread.append(th)
  #  join all thraeds  which are in thread
   for  thr in thread:
    thr.join()
```

```python
%%time
multi_file_download(l1)
```
